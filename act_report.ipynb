## Report: act_report

The analyses of the WeRateDogs dataset after cleaning lead to curious observations and further questions.


# WeRateDogs Popularity 

The retweets and favourite_count have a positive proportional relationship, the more retweets the more likes.

However this is not observed with the tweet frequency by year. Interestingly new tweets rose significantly between 2015 and 2016 but then substantially decreased in 2017.


Nevertheless the trend in terms of retweets and favourite_count was always rising across the years 2015, 2016, 2017.


Which can lead to the hypothesis that once the page reached a number of followers the decrease in new tweets in 2017 did not affect the popularity of the page with the retweets and favorite_count rising each year.
